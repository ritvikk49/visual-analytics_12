```python
import pandas as pd
import numpy as np
```

```python
df_original = pd.read_csv("spotify_dataset.csv") # Reading the actual dataset csv file
df = df_original.copy()
```

```python
df_original.info()
df_original.head(10)
```

```python
df = df[df.chart == 'top200']
# Modifying the dataset to have only the top200 chart and disassociate the viral50 chart 
```

```python
df = df.drop(['url','chart','trend'], axis=1)
# Droppping the url, chart and trend attribute columns. These attributes have no purpose in our case and increase size of the dataset.
```

```python
df = df[df['date'].str[-2:]=='01']
# Filtering rows with date '01' as we do not wish to consider all days and looking to bring down the size of dataset
```

```python
df = df[df['date'].str[:4] != '2017']
df = df[df['date'].str[:4] != '2018']
# Excluding records related to years 2017 and 2018 to decrease size of the dataset
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
# Changing the data type of the 'date' attribute to datetime64[ns]
```

```python
# Write data to a different CSV file. This will be used as the final dataset on which we will perform our analysis and visualisations
df.to_csv('transformed_dataset.csv', encoding='utf-8', index=False)
```
